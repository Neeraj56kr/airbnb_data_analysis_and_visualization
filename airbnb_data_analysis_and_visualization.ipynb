```python
import pandas as pd 
import numpy as  np 
import matplotlib.pyplot as plt
import seaborn as sns
```

#  Hendeling missing values 


```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

```python
df['NAME']
```

#  filling the missing value

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()},inplace=True )
```

```python
df.head()
```

#  remove the missing values NAME And host name

```python
df.dropna(subset=['NAME','host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

# remove unwanted columns license house_rules

```python
df=df.drop(columns=['house_rules','license'], errors='ignore')
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

#  remove  $  dollar signs and convert to float

```python
df['price'] =df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df['service fee'] =df['service fee'].replace('[\$]', '', regex=True).astype(float)
```

```python
df.head()
```

#  show and remove  values

```python
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df.head(100)
```

#  visulization 

#  what is distreibution of the listing price 


```python
plt.figure(figsize=(10,2))
sns.histplot(df['price'],bins=50,kde=True,color='green')
plt.title('distribution of listing price')
plt.xlabel("price $")
plt.ylabel("count OR frequency")
plt.show()
```

#  How are different room type distributed?

```python
df['room type']
```

```python
# 
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='room type',data=df,color='hotpink')
plt.title('room type distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

# how are listing distributed across different ?

```python
plt.figure(figsize=(10,7))
sns.countplot(y='neighbourhood group',data=df,color='lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title("Number of listings by neighbourhood group")
plt.xlabel('count')
plt.ylabel('neighbourhood group')
plt.show()
```

# what is the relationship b/w price and room type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price',hue='room type', data=df, palette='Set1')
```

# how has the number reviews change over with time

```python
reviews_over_time=df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(10,5))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of reviews Over time')
plt.xlabel('date')
plt.ylabel('Numberof reviews')
plt.show()
```
